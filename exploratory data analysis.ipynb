```python
#modifying the date and time into standard form
base_data['ScheduledDay'] = pd.to_datetime(base_data['ScheduledDay']).dt.date.astype('datetime64[ns]')
base_data['AppointmentDay'] = pd.to_datetime(base_data['AppointmentDay']).dt.date.astype('datetime64[ns]')
```

```python
base_data.head(5)
```

```python
#Scheduled day refers to the day that the appointment was made. This is the day that the patient and the hospital agreed upon for the patient to be seen by a doctor or other healthcare provider.
#Appointment day refers to the day that the patient actually goes to the hospital for their appointment. This may or may not be the same day as the scheduled day, depending on the hospital's availability and the patient's needs.
```

```python
# 5 is Saturday, 6 is Sunday ,0 is Monday

base_data['sch_weekday'] = base_data['ScheduledDay'].dt.dayofweek
```

```python
# create a bar plot of the count of the TARGET Variable 
base_data['NoShow'].value_counts().plot(kind='barh', figsize=(8, 6))
# add labels to the x-axis and y-axis
plt.xlabel("Count", labelpad=14)
plt.ylabel("Target Variable", labelpad=14)
# add a title to the plot
plt.title("Count of TARGET Variable per category", y=1.02);
```

```python


# create a pie chart of the count of the TARGET Variable
base_data['NoShow'].value_counts().plot(kind='pie', figsize=(8, 6))

# add labels to the pie chart
plt.legend(labels=base_data['NoShow'].value_counts().index, loc='best')

# add a title to the plot
plt.title("Count of TARGET Variable per category", y=1.02)

# show the plot
plt.show()
```

```python
# Group the ages
labels = ["{0} - {1}".format(i, i + 20) for i in range(1, 118, 20)]

base_data['Age_group'] = pd.cut(base_data.Age, range(1, 130, 20), right=False, labels=labels)
```

```python
#once we are having age_group column,we are dropping age column from base_data
base_data.drop(['Age'], axis=1, inplace=True)
```

```python
#having a look into the values of count of each columns and there count with respect to NoShow column
#BUT NOSHOW COLUMN IS NOT TAKEN INTO CONSIDERATION,IT IS DROPPED IN THIS CALCULATION
for i, predictor in enumerate(base_data.drop(columns=['NoShow'])):
    print('-'*10,predictor,'-'*10)
    print(base_data[predictor].value_counts()) 
    #for plotting graphs for all these predictors
    plt.figure(i)
    sns.countplot(data=base_data, x=predictor, hue='NoShow')
```

```python
# Convert 'Yes' values to 1 and 'No' values to 0 in the 'NoShow' column
base_data['NoShow'] = np.where(base_data.NoShow == 'Yes',1,0)
```

```python
base_data_dummies = pd.get_dummies(base_data)
base_data_dummies.head()
```
